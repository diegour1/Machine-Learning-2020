## Problema 2

```
# Arrays for the computations

TD = np.array([[2, 3, 0, 3, 7],[0, 5, 5, 0, 3],[5, 0, 7, 3, 3],[3, 1, 0, 9, 9],[0, 0, 7, 1, 3],[6, 9, 4, 6, 0]])
L = np.array([[5, 2, 3, 6, 4, 3]]).T
m , n = TD.shape
TD
```

```
# Problem 2(a)

S1 = TD.sum(axis=0).reshape((1,n))

# Joint Probability
M1 = (1/n)*(TD/S1)
print(M1.sum())
M1
```

```
# Problem 2(b)

# Probability P(T/D)
M2 = M1*n
M2
```

```
# Problem 2(d)

# Probability P(D)
P_D = (1/n)*np.ones(n).reshape(1,n)
P_D
```

```
# Problem 2(e)

# Probability P(T)
P_T = (M1.sum(axis=1)).reshape(m,1)
print(P_T.sum())
P_T
# 
```

```
# Problem 2(c)

# Probabilty P(D/T) Based on Bayes P(D/T) = P(T/D)*P(D)/P(T)
M3 = (M2*P_D)/P_T
print(M3.sum(axis=1).reshape((m,1)))
M3
```

```
# Problem 2(f)

# Expected value of l E[l], expected value of the lenght of the caracter for each document

E_l = (M2*L).sum(axis=0).reshape((1,n))
E_l
```

```
# Problem 2(g)

# Variance of the lenght l var(l), for each document

L_c = np.ones((m,n))*L
Var_l = (np.power((L_c-E_l),2)*M2).sum(axis=0)
Var_l
```
